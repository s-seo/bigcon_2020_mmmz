```python
# 중분류,  소분류, 4분류(?) 카테고리 긁어오기

import requests
from bs4 import BeautifulSoup as bs
import string
from pandas import DataFrame

body_cat = []
body_cat_s = []
body_cat_ss = []
code_m = []
code_s = []
code_ss = []
name_m = []
name_s = []
name_ss = []
for i in range(50000000,50000011):
    url = 'https://search.shopping.naver.com/category/category?catId=' + str(i)
    r = requests.get(url)
    html = r.content 
    ii = i-50000000
    soup = bs(html, "html.parser") 
    tag = soup.find_all("a", {'data-nclick':str("N=a:xa"+ list(string.ascii_lowercase[0:11])[ii] + ".m")})  
    tag_s = soup.find_all("a", {'data-nclick':str("N=a:xa"+ list(string.ascii_lowercase[0:11])[ii] + ".s")})
    tag_ss = soup.find_all("a", {'data-nclick':str("N=a:xa"+ list(string.ascii_lowercase[0:11])[ii] + ".ss")})
    
    for a in tag:
        body_cat.append('{\"name\":\"'  + a.text[0:-2] + '\",\"param\":[\"' + a["href"][23:31] + '\"]}')
        code_m.append(a["href"][23:31])
        name_m.append(a.text[0:-2])
    
    for c in tag_s:
        body_cat_s.append('{\"name\":\"'  + c.text + '\",\"param\":[\"' + c["href"][23:31] + '\"]}')
        code_s.append(c["href"][23:31])
        name_s.append(c.text)
        
    for b in tag_ss:
        body_cat_ss.append('{\"name\":\"'  + b.text + '\",\"param\":[\"' + b["href"][23:31] + '\"]}')
        code_ss.append(b["href"][23:31])
        name_ss.append(b.text)
```

```python
# 중분류, 소분류, 4분류(?) 카테고리 export to csv (R에서 작업하려고)

cat_m = {'code_m' : code_m, 'name_m':name_m}
DataFrame(cat_m).to_csv(r'C:\Users\baoro\Desktop\공모전\BCT_2020\data\cat_m.csv', 
             index = False,
             encoding='euc-kr')

cat_s = {'code_s':code_s, 'name_s':name_s}
DataFrame(cat_s).to_csv(r'C:\Users\baoro\Desktop\공모전\BCT_2020\data\cat_s.csv', 
             index = False,
             encoding='euc-kr')

cat_ss = {'code_ss':code_ss, 'name_ss':name_ss}
DataFrame(cat_ss).to_csv(r'C:\Users\baoro\Desktop\공모전\BCT_2020\data\cat_ss.csv', 
             index = False,
             encoding='euc-kr')
```
